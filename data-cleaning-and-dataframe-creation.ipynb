```python
import pandas as pd
import numpy as np
import src.index as i
```

```python
passengers = pd.read_csv("Inputs/passengers.csv")
```

```python
passengers = passengers[passengers["TRA_MEAS"] == "Passengers carried"]
```

```python

passengers = pd.pivot_table(passengers, values='Value', index=['GEO'],columns="TIME",aggfunc=np.sum)
```

```python
passengers.index
```

```python
passengers.insert(0,"Traffic_type","passengers")
```

```python
passengers.head()
```

```python
passengers.index
```

```python
passengers.index = i.valueschanger(["Germany (until 1990 former territory of the FRG)"], passengers.index)
```

```python
passengers.head()
```

```python
passengers = passengers.replace(":", 0)
passengers.reset_index()
passengers.head()
```

```python

passengers = passengers.reset_index()
passengers = passengers.rename(columns={"index":"Country"})
passengers.head()
```

```python
tones = pd.read_table("Inputs/Tones.tsv")
```

```python
tones.insert(1,"Traffic_type","Cargo")
```

```python
tones = tones.rename(columns={"freq,unit,tra_meas,schedule,tra_cov,geo\TIME_PERIOD":"Country"})
```

```python
tones.Country = tones.Country.apply(i.countrychanger)
```

```python
passengers.head()
```

```python
tones.head()
```

```python
final = pd.concat([passengers,tones])
```

```python
final.columns
```

```python
final = final.groupby(by=["Country","Traffic_type"]).agg(sum)
```
